```python
# Creating a subset of dataframe
df = df_ords_prods_merged[:1000000] 
```

```python
# Defining price conditions 
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Using the defined price-label function on the dataframe
df['price_label'] = df.apply(price_label, axis=1)
```

```python
# Checking values in th e new column
df.value_counts()
```

```python
# Checking for the most expensible products in the dataframe
df['prices'].max()
```

```python
#If-Statements with the loc() Function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
#Using the price_range _loc function
df['price_range_loc'].value_counts(dropna=False)
```
