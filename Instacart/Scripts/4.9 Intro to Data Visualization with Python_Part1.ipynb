# Answer 5

```python
# Data quality and consistency checks  - Mixed- type data
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```

```python
# Data quality and consistency checks  - Missing values
df_customers.isnull().sum()
```

```python
# Data quality and consistency checks  - Duplicates
df_customers_dups = df_customers[df_customers.duplicated()]
```

```python
df_customers_dups.shape
```

```python
# We have no mixed-type data, no duplicates, just the 11259 first _name missing records, so i will permanently drop that column
df_customers.drop(columns = ['First Name'], inplace = True)
```
